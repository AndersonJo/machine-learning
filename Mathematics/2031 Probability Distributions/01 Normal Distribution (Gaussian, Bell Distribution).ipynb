# Normal Distribution

일반적으로 수학자들은 Normal distribution이라고 부르고, 물리학자들의 경우 Gaussian distribution 이라고 하며, <br>
사회 과학자들은 Bell Curve 라고도 부릅니다. 

Standard normal distribution은 $ \mu = 0 $ 그리고 $ \sigma^2 = 1 $ 값을 갖습니다. <br>
$ \mu $ 는 location parameter라고도 하며 center가 되는 위치를 나타내며, <br>
$ \sigma $는 scale parameter라고도 하며 얼마나 퍼져있는지를 나타냅니다.


### $$ P(x) = \frac{1}{\sqrt{2\pi \sigma^2}} e^{ - \frac{(x-\mu)^2}{2 \sigma^2} } $$

```python
%pylab inline
import numpy as np
import scipy.stats
```

### Normal Distribution with Numpy

```python
data = np.random.normal(size=30000)

title('Normal Distribution with Numpy')
grid()
hist(data, bins=100, normed=True)

print()
```

# 집접 만든 Normal Distribution

```python
def normal_distribution(x, location=0., scale=1.):    
    return 1./(scale * np.sqrt(2.*np.pi)) * np.exp(-(x-location)**2./(2. * scale**2))

x = np.arange(-6, 6, 0.01) 
grid()
plot(x, normal_distribution(x, location=0, scale=1), label='location:0, scale:1')
plot(x, normal_distribution(x, location=2, scale=1), label='location:2, scale:1')
plot(x, normal_distribution(x, location=0, scale=3), label='location:2, scale:3')
legend(loc=2)

print(np.max(data))
```
